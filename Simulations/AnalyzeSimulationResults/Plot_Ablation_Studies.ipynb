# Synthetic Low-Dim Experiment - Complexity Regularization $\sigma^2$ Value vs Performance

## Synthetic Low-Dimensional Setting

```python
df_kernel_alternative_proxy = pd.read_pickle(r"../Results/Kernel_Alternative_Proxy_Synthetic_Experiment1.pkl")
df_kernel_alternative_proxy = df_kernel_alternative_proxy.loc[(df_kernel_alternative_proxy['scale_data'] == True) & (df_kernel_alternative_proxy['stage1_perc'] == 0.5)]
print('Shape: ', df_kernel_alternative_proxy.shape)

df_kernel_alternative_proxy_sigmasquare_0 = df_kernel_alternative_proxy.loc[df_kernel_alternative_proxy['sigma_square'] == 0.0]
df_kernel_alternative_proxy_sigmasquare_1 = df_kernel_alternative_proxy.loc[df_kernel_alternative_proxy['sigma_square'] == 1.0]
df_kernel_alternative_proxy_sigmasquare_2 = df_kernel_alternative_proxy.loc[df_kernel_alternative_proxy['sigma_square'] == 2.0]
df_kernel_alternative_proxy_sigmasquare_3 = df_kernel_alternative_proxy.loc[df_kernel_alternative_proxy['sigma_square'] == 3.0]

df_kernel_alternative_proxy_sigmasquare_0['Complexity Regularization'] = '$\sigma^2 = 0$'
df_kernel_alternative_proxy_sigmasquare_1['Complexity Regularization'] = '$\sigma^2 = 1$'
df_kernel_alternative_proxy_sigmasquare_2['Complexity Regularization'] = '$\sigma^2 = 2$'
df_kernel_alternative_proxy_sigmasquare_3['Complexity Regularization'] = '$\sigma^2 = 3$'

df_results = pd.concat([df_kernel_alternative_proxy_sigmasquare_0,
                        df_kernel_alternative_proxy_sigmasquare_1,
                        df_kernel_alternative_proxy_sigmasquare_2,
                        df_kernel_alternative_proxy_sigmasquare_3,
                      ],
                      axis = 0)

SetPlotRC()
sns.boxplot(data=df_results[["Complexity Regularization", "Data_Size", "Causal_MSE"]], x='Data_Size', y='Causal_MSE', hue='Complexity Regularization')
plt.xlabel("Data Size", fontsize=10)
plt.ylabel("Mean Squared Error", fontsize=10)
plt.title("Synthetic Low-Dimensional Data - Ablation Study on $\sigma^2$", fontsize=1)
plt.grid(linewidth=1.5)
# plt.legend(loc = 'upper center')
ApplyFont(plt.gca(), xlabel_text_size = 15.0, ylabel_text_size = 15.0, title_text_size = 15.0, ticks_text_size = 15,)
plt.savefig('Figures/Alternative_Proxy_ATE_SyntheticLowD_ComplexityRegularizationAblation.pdf', format='pdf', bbox_inches = 'tight')
plt.show()
```
